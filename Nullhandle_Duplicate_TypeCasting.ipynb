## CHECKING NULL VALUES (NaN) / MISSING VALUES


```python
import pandas as pd 
```

```python
df=pd.read_csv('Book1.csv')
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

## HANDLING MISSING VALUE USING BFILL AND FFILL/PAD

```python
df1=df.fillna(method='bfill')
```

```python
df1
```

```python
df2=df.fillna(method='ffill')
```

```python
df2
```

```python
df3=df.fillna(method='pad')
```

```python
df3
```

## HANDLING MISSING VALUE FOR A PARTICULAR COLUMN

```python
df4=df['std_address'].fillna(method='ffill')
```

```python
df4
```

```python
df5=df.fillna({'std_address':'NO_ADDRESS','std_dept':'NO_DEPT'})
```

```python
df5
```

## FINDING DUPLICATES ROW

```python
#checking respective to all the column
df6=df.duplicated()
```

```python
df6
```

```python
df7=df.duplicated(subset=['std_name','std_number'])
```

```python
df7
```

## DISPLAYING FIRST,LAST,ALL DUPLICATE ROWS

```python
df8=df.loc[df.duplicated(subset=['std_name'],keep=False)]
```

```python
df8
```

```python
df9=df.loc[df.duplicated(subset=['std_name'],keep='first')]
```

```python
df9
```

## DROPPING DUPLICATE VALUES

```python
df10=df.drop_duplicates()
```

```python
df10
```

```python
df11=df.drop_duplicates(subset=['std_name','std_number'])
```

```python
df11
```

```python
df12=df.drop_duplicates(subset=['std_name','std_number'],keep='last')
```

```python
df12
```

```python
df13=df.drop_duplicates(subset=['std_name','std_number'],keep=False)
```

```python
df13
```
